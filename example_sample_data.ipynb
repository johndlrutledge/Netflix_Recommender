```python
import pandas as pd
import numpy as np
```

```python
# read into dataframe
# movie_user_rating_date1.txt is from original comined_data_1.txt file, but with movie # inserted
#   before each user,rating,date (as opposed to original which had movie as a single line before all users/ratings)
#   to make it easier to load into dataframe
data1 = pd.read_csv('movie_user_rating_date1.txt', sep=',', header=0, names=['Movie','User','Rating','Date'])
data1.head()
```

```python
# random sample, adjust n as needed for sample size
data_small_date = data1.sample(n=20000)
```

```python
# create nested dictionary: {user: {movie: [rating, date]}, {movie: [rating, date]} ....}
rating_dict = {}
for index in range(0,len(data_small_date)-1):
    user = data_small_date.iloc[index].at['User']
    movie = data_small_date.iloc[index].at['Movie']
    date = data_small_date.iloc[index].at['Date']
    rating = data_small_date.iloc[index].at['Rating']
    if user in rating_dict.keys():
        rating_dict[user][movie] = [rating, date]
    else:
        rating_dict[user] = {movie: [rating, date]}
```
